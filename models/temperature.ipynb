### You are given a CSV file containing daily temperature readings for 5 cities over 365 days. Tasks need to perform:

a) Read the file using Pandas and convert the temperature data into a NumPy array.

```python
import pandas as pd 

import numpy as np



data = pd.read_csv("city_temperatures.csv")

print(data.head())

print(data.shape)
```

b) Using NumPy slicing and indexing, extract:



i. The temperature of the 3rd city for the last 100 days



ii. The hottest temperature recorded for each city

```python
# select Date and Kolkata columns and take the last 100 rows

city3data = data[["Date", "Kolkata"]].iloc[-100:]

print(city3data)
def getHottestTemp(temp_column):

    return temp_column.max()





for city in data.columns[1:]:  

    print(f"The hottest temperature in {city} is:", getHottestTemp(data[city]))
```

c) Use NumPy filtering to find all days where any city had a temperature above 45Â°C.

```python
temps = data.iloc[:, 1:].to_numpy()



result = data[(temps > 45).any(axis=1)]



print(result)
```

d) Use array splitting to divide the year into seasons (four equal parts) and compute

the average temperature for each season.

```python
# Split the temperature data into 4 equal parts (seasons)
seasons = np.array_split(temps, 4)

# Compute the average temperature for each season
season_averages = [np.mean(season) for season in seasons]

# Print the average temperature for each season
for i, avg in enumerate(season_averages, 1):
    print(f"Average temperature for season {i}: {avg:.2f}")
```
